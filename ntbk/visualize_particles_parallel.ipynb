```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
indir = "/Users/jgra0019/Documents/codes/uw3-dev/advection-tests/src"
bef_adv_prefix = "bef_adv"
aft_adv_prefix = "aft_adv"
aft_res_prefix = "aft_res"

nranks = 4
```

```python
# plot before advection 
num_particles = 0

fig, ax = plt.subplots(dpi = 300)
for i in range(nranks):
    fname = f"{indir}/{bef_adv_prefix}_r{i}.csv"
    df = pd.read_csv(fname, header = None)
    df.columns = ["X", "Y"]
    ax.scatter(df["X"], df["Y"], s = 5, c = f"C{i}", label = f"Rank {i}")
    #num_particles += len(df)
    num_particles += (~df["X"].isna()).sum()
    print(f"Rank: {i}: {(~df.X.isna()).sum()}")

ax.set_aspect("equal")
ax.set_xlim([-0.1, 4.1])
ax.set_ylim([-0.1, 1.1])
ax.legend(ncol = 2)
print(num_particles)
```

```python
# plot after advection 
import numpy as np

num_particles = 0

fig, ax = plt.subplots(dpi = 300)
for i in range(nranks):
    fname = f"{indir}/{aft_adv_prefix}_r{i}.csv"
    df = pd.read_csv(fname, header = None)
    df.columns = ["X", "Y", "R0"]
    for j, row in df.iterrows():
        if row["R0"] == 0: 
            color   = "C0"
            size    = 20
            zorder  = 0
        elif row["R0"] == 1:
            color   = "C1"
            size    = 10
            zorder  = 1
        elif row["R0"] == 2:
            color   = "C2"
            size    = 5
            zorder  = 2
        elif row["R0"] == 3:
            color   = "C3"
            size    = 2
            zorder  = 3
            #print(row)
        
        ax.scatter(row["X"], row["Y"], s = size, c = color, zorder = zorder)

    num_particles += (~df["X"].isna()).sum()
    print(f"Rank: {i}: {(~df.X.isna()).sum()}")

ax.set_aspect("equal")
ax.set_xlim([-0.1, 4.1])
ax.set_ylim([-0.1, 1.1])
#ax.legend(ncol = 2, loc = "lower right")
print(num_particles)
```

```python
# plot after restoration 
num_particles = 0

fig, ax = plt.subplots(dpi = 300)
for i in range(nranks):
    fname = f"{indir}/{aft_res_prefix}_r{i}.csv"
    print(fname)
    df = pd.read_csv(fname, header = None)
    df.columns = ["X", "Y", "R0"]
    
    for j, row in df.iterrows():
        if row["R0"] == 0: 
            color   = "C0"
            size    = 20
            zorder  = 0
        elif row["R0"] == 1:
            color   = "C1"
            size    = 10
            zorder  = 1
        elif row["R0"] == 2:
            color   = "C2"
            size    = 5
            zorder  = 2
        elif row["R0"] == 3:
            color   = "C3"
            size    = 2
            zorder  = 3
        ax.scatter(row["X"], row["Y"], s = size, c = color, zorder = zorder)
    
    num_particles += (~df["X"].isna()).sum()
    print(f"Rank: {i}: {(~df.X.isna()).sum()}")

ax.set_xlim([-0.1, 4.1])
ax.set_ylim([-0.1, 1.1])
ax.set_aspect("equal")
ax.legend(ncol = 2)
print(num_particles)
```
